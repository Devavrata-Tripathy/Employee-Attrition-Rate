```python
df.drop(df.iloc[:,15:], inplace = True, axis = 1)
df.head(5)
```

```python
x = df.iloc[:,1:].values
print(x)
```

```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
l_x = LabelEncoder()
x[:,0] = l_x.fit_transform(x[:,0])
x[:,1] = l_x.fit_transform(x[:,1])
x[:,2] = l_x.fit_transform(x[:,2])
x[:,3] = l_x.fit_transform(x[:,3])
x[:,4] = l_x.fit_transform(x[:,4])
#print(x)
ohc = OneHotEncoder()
x = pd.DataFrame(ohc.fit_transform(x).toarray())
print(x)
#x = np.array(columnTransformer.fit_transform(data), dtype = np.str) 
```

```python
df1 = pd.read_csv('Train.csv')
df1.drop(df1.iloc[:,14:15], inplace = True, axis = 1)
df1.drop(df1.iloc[:,:8], inplace = True, axis = 1)
df1.head(5)
```

```python
df2 = df1.join(x)
df2.head(5)
```

```python
y_t = df2.iloc[:, 14].values
df2.drop(df2.iloc[:,14:15], inplace = True, axis = 1)
x_t = df2.iloc[:,:].values


from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.nan , strategy = 'mean')
imp = imp.fit(x_t)
x_t = imp.transform(x_t)
print(y_t)
print(x_t.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_t,y_t)
```
