```python
#Importing the essential libraries...

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the dataset from Train.csv file
df = pd.read_csv('Train.csv')
df.head(5)
```

```python
#Some basic and essential information regarding the dataset 
print(df.info())
df.describe()
```

```python
#Using the heatmap for plotting the various correlations
#As correlations play an extremely important role 
#the correlation matrix can help to identify the strongly correlated features

plt.figure(figsize = (16, 6))
cor = df.corr()
sns.heatmap(data = cor, annot = True, linewidth = 0.5 )
```

```python
#Plotting the distribution of response vector i.e Attrition_rate
#The plot shows the skewness of the dataset 

plt.figure(figsize = (16, 6))
x1 =  df.iloc[:,-1].values
sns.distplot(x1, norm_hist = True)
```

```python
#Lets plot the various categorical features
#using the countplot to count the no of occurances and get an overview of data

fig2, ax = plt.subplots(nrows = 2,ncols=2, figsize = (30, 20))
sns.countplot(data = df, x = 'Hometown', ax = ax[0][0])
sns.countplot(data = df, x = 'Unit', ax = ax[0][1])
sns.countplot(data = df, x = 'Gender', ax = ax[1][0])
sns.countplot(data = df, x = 'Decision_skill_possess', ax = ax[1][1])
ax[0][1].set_xticklabels(labels = df.iloc[:,6].values ,rotation=-90)
```

```python
#Lets see if Gender and Relationship has any effect ...

plt.figure(figsize = (16, 6))
g = sns.FacetGrid(data = df, col = 'Gender', row = 'Relationship_Status', height = 6, aspect = 16/6)
g.map(sns.distplot, 'Attrition_rate')
```

```python
#Plotting the relation between education level and Attrition_rate
#lets see the relation ....

plt.figure(figsize = (12, 6))
sns.boxplot(x='Education_Level',y='Attrition_rate', data=df)
```

```python
#I have plotted the regression plot for the various features and more importantly 
#for the 7 unknown numerical variables VAR1 VAR2 VAR3 VAR4 VAR5 VAR6 VAR7 
#All of them look kind of similar 

fig, axs = plt.subplots(nrows = 5, ncols=2 , figsize = (15, 50))
sns.regplot(x='Work_Life_balance', y='Attrition_rate', data=df, ax=axs[0][0], color = 'orange', marker = 'x')
sns.regplot(x='Post_Level', y='Attrition_rate', data=df, ax=axs[0][1])
sns.regplot(x='Education_Level',y='Attrition_rate', data=df, ax=axs[1][0])
sns.regplot(x='VAR1', y='Attrition_rate', data=df, ax=axs[1][1])
sns.regplot(x='VAR2', y='Attrition_rate', data=df, ax=axs[2][0])
sns.regplot(x='VAR3', y='Attrition_rate', data=df, ax=axs[2][1])
sns.regplot(x='VAR4', y='Attrition_rate', data=df, ax=axs[3][0])
sns.regplot(x='VAR5', y='Attrition_rate', data=df, ax=axs[3][1])
sns.regplot(x='VAR6', y='Attrition_rate', data=df, ax=axs[4][0])
sns.regplot(x='VAR7', y='Attrition_rate', data=df, ax=axs[4][1])
```

```python
#I have plotted the distribution of the 7 unknown variables

fig, axs = plt.subplots(nrows = 4, ncols=2 , figsize = (15, 50))
var1 = df.iloc[:,-8]
var2 = df.iloc[:,-7]
var3 = df.iloc[:,-6]
var4 = df.iloc[:,-5]
var5 = df.iloc[:,-4]
var6 = df.iloc[:,-3]
var7 = df.iloc[:,-2]

sns.distplot(var1, ax =axs[0][0], kde = False)
sns.distplot(var2, ax =axs[0][1])
sns.distplot(var3, ax =axs[1][0])
sns.distplot(var4, ax =axs[1][1], kde = False)
sns.distplot(var5, ax =axs[2][0])
sns.distplot(var6, ax =axs[2][1])
sns.distplot(var7, ax =axs[3][0])
```

```python
#The Age distribution looks fair ... it has somewhat good negative correlation 

plt.figure(figsize = (16, 6))
sns.barplot(x='Age',y='Attrition_rate', data=df, palette = 'husl')
plt.xticks(rotation = -90)
```

```python
plt.figure(figsize = (26, 6))
sns.relplot( data = df , x = 'Unit', y = 'Attrition_rate', hue = 'Age', size = 'Gender')
plt.xticks(rotation = -90)
```
